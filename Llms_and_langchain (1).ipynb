1.What are Large Language Models (LLMs) and how do they function?

ans-

    Large Language Models (LLMs) are advanced AI systems trained on vast amounts of text data to understand, generate, and manipulate human language. Examples include GPT models.

    How they function:

    Built using transformer architecture
    Use self-attention mechanisms to understand context
    Trained on massive datasets using unsupervised learning
    Predict the next word/token based on previous context
    Fine-tuned for specific tasks (chatbots, translation, etc.)

6.Implement a basic Langchain pipeline using OpenAI’s LLM to answer
questions based on a user input prompt

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
import os

# Step 1: Set your API key
os.environ["GEIMINI_API_KEY"] = "GEMINI_API_KEY"

# Step 2: Initialize Gemini model
llm = ChatGoogleGenerativeAI(
    model="gemini-pro",
    temperature=0.7
)

# Step 3: Create prompt template
prompt = ChatPromptTemplate.from_template(
    "Answer the following question clearly:\n{question}"
)

# Step 4: Create chain (pipeline)
chain = prompt | llm

# Step 5: Take user input
user_question = input("Enter your question: ")

# Step 6: Run the chain
response = chain.invoke({"question": user_question})

# Step 7: Print output
print("\nAnswer:")
print(response.content)
```

7.Integrate Langchain with a third-party API (e.g., weather, news) and show
how responses can be generated via LLMs

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
import requests
import os

# Step 1: Set API keys
os.environ["GOOGLE_API_KEY"] = "GEMINI_API_KEY"
WEATHER_API_KEY = "YOUR_WEATHER_API_KEY"

# Step 2: Function to fetch weather data
def get_weather(city):
    url = f"http://api.weatherapi.com/v1/current.json?key={WEATHER_API_KEY}&q={city}"
    response = requests.get(url)
    return response.json()

# Step 3: Initialize Gemini LLM
llm = ChatGoogleGenerativeAI(
    model="gemini-1.5-flash",
    temperature=0.7
)

# Step 4: Create prompt template
prompt = ChatPromptTemplate.from_template(
    """You are an assistant.
    Based on the following weather data, explain the current conditions in a simple way.

    Weather Data:
    {weather_data}

    User Question:
    {question}
    """
)

# Step 5: Create chain
chain = prompt | llm

# Step 6: User input
city = input("Enter city: ")
question = input("Ask a question about the weather: ")

# Step 7: Fetch API data
weather_data = get_weather(city)

# Step 8: Generate response using LLM
response = chain.invoke({
    "weather_data": weather_data,
    "question": question
})

# Step 9: Output
print("\nAnswer:")
print(response.content)
```

retrieves answers from it.

```python
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex
from llama_index.llms.google_genai import GoogleGenAI
import os

# Step 1: Set API key
os.environ["GOOGLE_API_KEY"] = "GEMINI_API_KEY"

# Step 2: Load documents from local directory
documents = SimpleDirectoryReader("sample.txt").load_data()

# Step 3: Initialize LLM (Gemini)
llm = GoogleGenAI(model="gemini-1.5-flash")

# Step 4: Create index
index = VectorStoreIndex.from_documents(documents)

# Step 5: Create query engine
query_engine = index.as_query_engine(llm=llm)

# Step 6: Ask a question
query = input("Enter your question: ")

response = query_engine.query(query)

# Step 7: Print response
print("\nAnswer:")
print(response)
```
